```python
t = ('Norway', 3, 4.953)
t
```

```python
type(t)
```

```python
t[0]
```

```python
t[1]
```

```python
t[2]
```

- we can determine the number of elements in the tuple using the built-in
len function

```python
len(t)
```

```python
for item in t:
    print(item)
```

- we can concatenate tuples using the plus operator

```python
t + (334443,2)
```

- or repeat using the multiplication operator

```python
t * 3
```

### <font color=lightGreen>String Formatting</font></h3>

with positional arguments

```python
"Current position {latitude} {longitude}".format(latitude="60N", longitude="50E")
```

It's possible to index in the sequence using square brackets
inside the replacement field


```python
"Galactic position x={pos[0]}, y={pos[1]}, z={pos[2]}".format(pos=(65.2, 23.1, 82.2))
```


you can even access module attributes

```python
import math
"Math constants: pi={m.pi}, e={m.e}".format(m=math)
```


Format strinngs gives control over field alingment and floating point formating,
in the following example we're displaying the same values using only 3 decimal places.


```python
"Math constants: pi={m.pi:.3f}, e={m.e:.3f}".format(m=math)
```

```python
value = 4*20
'The value is {value}'.format(value=value)
```

### <font color=lightGreen>Slicing</font></h3>

Extended form of indexing for referring to a portion of a list or other sequence
Syntax:
- `a_list[start:stop]`

```python
s = [100, 101, 102, 103, 104, 105, 106]
s[1:3] #from index one to four
```

```python
s[1:-1] #from index one to last
```

```python
s[2:] #another way, from index two to last
```

```python
s[:2] #from first to index two
```

```python
s[:] #all elements
```

`= listName[:]` is a good way to copy a list avoiding pointing to the same object

```python
r = s[:]
r is s
```

```python
r == s
```

```python
u = s.copy()
u is s
```

-   A set of operations that a type must support to implement the protocol
-   Do not need to be defined as interfaces or base classes
-   Types only need to provide functioning implementations

<font color=lightGreen>

|<center>Protocol</center>|<center>Implementing collections</center>|
|--------|------------------------|
|Container|`str, list, dict, range, tuple, set, bytes`|
|Sized|`str, list, dict, range, tuple, set, bytes`|
|Iterable|`str, list, dict, range, tuple, set, bytes`|
|Sequence|`str, list, range, tuple, bytes`|
|Mutable Sequence|`list`|
|Mutable Set|`set`|
|Mutable Mapping|`dict`|

</font>
