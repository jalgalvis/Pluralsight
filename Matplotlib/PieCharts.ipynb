```python
import numpy as np
import matplotlib.pyplot as plt
params = {"ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.edgecolor" : "w",
          "text.color" : 'w'}
plt.rcParams.update(params)
```

## <center><font color=tomato>Pie Charts</font></center>
-   Like the stacked bar chart, represents parts of a whole
-   Displays only a single dataset
-   Values are percentages
-   Will always add up to 100
-   The entire dataset is represented as a circle
-   The values are wedges
-   The size of each wedge is proportional to the total of the values
-   Each wedge/value is represented by a slice of the pie/dataset

The pie function will place the first wedge on the right side of the pie parallel to the X-axis and display the remaining wedges moving counter-clockwise.

```python
wedges = np.array([8, 20, 20, 52])
plt.pie(x=wedges);
```

Each wedge can also be labeled by passing an iterable to the `labels` keyword argument.
By default, the labels are drawn parallel to the X-axis.


```python
plt.pie(x=wedges, labels='ASDF');
```

The `rotatelabels` keyword argument as set to True will rotate each label to the angle of the slice

```python
plt.pie(x=wedges, labels=['Blue', 'Orange', 'Green', 'Red'], rotatelabels=True);
```

The `labeldistance` keyword argument controls how far from the center of the pie the labels are drawn.
A value of 1 will draw the labels on the edge of the pie. The default value is 1.1

```python
plt.pie(x=wedges, labels=['Blue', 'Orange', 'Green', 'Red'], labeldistance=1.5);
```

Sometimes it can be difficult to determine the relative sizes of the wedges just by looking at the them if the values are very similar.
In this case, the pie function will automatically calculate the percentages of each wedge and display them inside of the wedge.
To do this, include the `autopercent` keyword argument and give it a format string.

In this example, I am telling Matplotlib to display the value as a floating point percentage with two decimal places.
It will be followed by the percent sign.
The two percent signs at the end of the format string is the sequence for an escaped percent sign to be displayed.

```python
plt.pie(x=wedges, autopct='%.2f%%');
```

<font color=mediumTurquoise>Exploding</font> a wedge

Small wedges can be seen easily if they are offset from the rest of the pie
-   The `explode=` keyword argument is a list of vales, one for each wedge
-   The values instruct `matplotlib` how much to offset the wedges, as a fraction of the radius.

```python
plt.pie(x=wedges, explode=[0.5, 0, 0, 0]);
```

```python
precip_2018 = [3.92,
             1.7000000000000002,
             4.800000000000001,
             5.35,
             5.119999999999999,
             1.98,
             4.919999999999999,
             6.619999999999999,
             7.9700000000000015,
             4.579999999999999,
             7.1,
             6.2299999999999995]
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
       'Oct', 'Nov', 'Dec'])
plt.pie(x=precip_2018, labels=months, startangle=90, autopct='%.1f%%')
```

```python
_, years, percentages = plt.pie(x=precip_2018, autopct='%.1f%%', labels=months)
years[0].set_fontweight('bold')
percentages[2].set_fontstyle('italic')
```

```python
explotions = np.zeros(12)
explotions[np.argmin(precip_2018)] = 0.4
explotions[np.argmax(precip_2018)] = 0.6
plt.pie(x=precip_2018, labels=months, explode=explotions);
```
