The color cycle is of type `cycler` found in the `cycler` module, so I'll need to import it.

This will be the keyword argument color, and the value will be a list of the hexadecimal values that I'm selecting from Adobe color.
To set the `cycler` as the c`olor cycle`, I'll use the `rc` function in the `matplotlib.pyplot` module.

```python
from cycler import cycler
my_cycler = cycler(color=['#9038FF', '#5233E8', '#4558FF', '#3375E8', '#38B5FF'])
plt.rc('axes', prop_cycle=my_cycler)
for offset in range(10):
    plt.plot(x, y + offset);
```

## <center><font color=tomato>Line Style</font></center>
-   Lines created with the plot() function
-   Use the `linestyle=` keyword argument
-   Linestyles have names and abbreviations:
    -   Name: 'solid', abbreviation: '-'
    -   Name: 'dashed', abbreviation: '--'
    -   Name: 'dashdot', abbreviation: '-.'
    -   Name: 'dotted', abbreviation: ':'
-   The 'None' style
    -   Removes the line completely
    -   Abbreviation is single space or empty string

### <font color=lightGreen>Dashes</font>
-   The `dash_capstyle=` keyword arguemnt
    -   Butt (default)
    -   Round
    -   Projecting
-   The `dash_joinstyle=` keyword arguemnt
    -   Round (default)
    -   Miter
    -   Bevel

### <font color=lightGreen>Line width</font>
-   The `linewidth=` keyword argument


```python
styles = ['solid', 'dashed', 'dashdot', 'dotted', 'None']

for index, style in enumerate(styles):
    plt.plot(np.arange(10) + index, linestyle= style);
```

```python
abbreviations = ['-', '--', '-.', ':', '']
for index, style in enumerate(abbreviations):
    plt.plot(np.arange(10) + index, linestyle= style);
```

```python
plt.plot(np.arange(10), linestyle= (0, (1, 10, 5, 1)));
```

```python
plt.plot(np.arange(10), linestyle='--', linewidth=10)
plt.plot(np.arange(10) + 1, linestyle='--', linewidth=10, dash_capstyle='round');
```

## <center><font color=tomato>Markers</font></center>
The `marker=` keyword argument

| keyword | marker|
|-|-|
| Lowercase ‘o’ | circle|
| Period ‘.’ | point|
| Lowercase ‘s’ | square|
| Lowercase ‘v’ | triangle down|
| Caret ‘*’ | triangle up|
| Less than ‘<‘ | triangle left|
| Greater than ‘>’ | triangle right|
| Asterisk ‘*’ | five sided star|
| Plus ‘+’ | plus with thin lines|
| Capital ‘P’ | plus with thick lines|
| Lowercase ‘p’ | pentagon|
| Eight ‘8’ | octagon|

```python
for index, symbol in enumerate(list('osv^<>*Pp8')):
    plt.scatter([0 + index], [0], marker=symbol);
```

```python
points = np.array(list(zip(np.random.uniform(size=100), np.random.uniform(size=100))))
```

```python
quadrants = {
    1: [],
    2: [],
    3: [],
    4: []
}
```

```python
for point in points:
    if point[0] >= 0.5 and point[1] >= 0.5: quadrants[1].append(point)
    if point[0] < 0.5 and point[1] >= 0.5: quadrants[2].append(point)
    if point[0] < 0.5 and point[1] < 0.5: quadrants[3].append(point)
    if point[0] >= 0.5 and point[1] < 0.5: quadrants[4].append(point)
```

```python
for k in quadrants.keys():
    quadrants[k] = np.array(quadrants[k])
```

```python
plt.scatter(quadrants[1][:,0], quadrants[1][:,1], marker=(3, 0, 75), s=np.ones(len(quadrants[1])) * 100)
plt.scatter(quadrants[2][:,0], quadrants[2][:,1], marker=(3, 0, 165), s=np.ones(len(quadrants[2])) * 100, c='C0')
plt.scatter(quadrants[3][:,0], quadrants[3][:,1], marker=(3, 0, 255), s=np.ones(len(quadrants[3])) * 100, c='C0')
plt.scatter(quadrants[4][:,0], quadrants[4][:,1], marker=(3, 0, 345), s=np.ones(len(quadrants[4])) * 100, c='C0');
```

## <center><font color=tomato>Legends</font></center>
-   Generated by the `legend()` function
-   Legends are built using labeled lines with the `label=` keyword argument
-   Also could be done using handles, passing the lines and labels to the `legend()` function

### <font color=lightGreen>Location</font>
Will be inferred by default to a position that does not obscure the visualization

The `loc=` keyword argument
-   Explicit legend location
-   Can be faster for complex visualizations


```python
years = range(2015, 2018)
data = {}
for y in years:
    data[y] = list(np.random.randint(low=1, high=20, size=12))
```

```python
for key in sorted(data.keys()):
    plt.plot(data[key], label=str(key))
plt.legend();
```

To call out 2015, I'm going to store the handle of each line.
The output of the plot function is a list with a single handle for the line.
So I'll need to pull it out of the list using the 0 index.
Now I can change the width of that line with the `set_lw` function and change the style of the line to dotted with a `set_ls` function.
Notice that the legend updates to reflect the wider dotted line

```python
handles = []
for key in sorted(data.keys()):
    line = plt.plot(data[key], label=str(key))[0];
    handles.append(line)
handles[0].set_lw(5)
handles[0].set_ls(':')
plt.legend();
```

### <font color=lightGreen>Legend</font> appearance
Arguments in `legend()`
-   Put the legend along the bottom with `loc=8` which means lower center,
-   and, all the values in a single row with `ncol=3`
-   change the background color with `facecolor='#FFCCCC'`
-   change the fontsize with `fontsize='large'`
-   add shadow with `shadow=True`
-   add title with `title='Years'`


```python
handles = []
for key in sorted(data.keys()):
    line = plt.plot(data[key], label=str(key))[0];
    handles.append(line)
handles[0].set_lw(5)
handles[0].set_ls(':')
plt.legend(loc=8, ncol=3, facecolor='#FFCCCC', fontsize='large', shadow=True, title='Years');
```

The pie function does not let assign labels to wedges without displaying the labels themselves.
So we can't use that method to tell the legend which wedges to label.
However, we can give a list of the handles to the wedges and pass them explicitly to the legend function.
The wedges are the first element in the output of the pie function.
So I'll store those and then pass them to the legend function, along with the month names.

```python
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
       'Oct', 'Nov', 'Dec'])
wedges, _ = plt.pie(data[2015]);
plt.legend(wedges, months);
```

This creates a legend, but it's covering a significant portion of the pie.
I can solve this by using the bbox_to_anchor keyword argument.
with a tuple with an X and Y coordinate.

```python
wedges, _ = plt.pie(data[2015]);
plt.legend(wedges, months, bbox_to_anchor=(1.0, 1.0));
```

Adding handles also works for bar Charts and scatter plots!!

```python
group_width = 0.9
offsets = [-group_width / 3, 0, group_width / 3]
for index, key in enumerate(sorted(data.keys())):
    plt.barh(np.arange(len(data[key])) - offsets[index], data[key], height=group_width / 3, label=str(key));
plt.yticks(ticks=np.arange(len(months)), labels=months);
plt.legend();
```
