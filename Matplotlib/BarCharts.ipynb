```python
import numpy as np
import matplotlib.pyplot as plt
params = {"ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.edgecolor" : "w"}
plt.rcParams.update(params)
```

## <center><font color=tomato>Bar chart</font></center>
    -   Similar to a line chart
    -   Represents individual values with bars
    -   Taller bars represent greater values
    -   Generated with `bar()` function
    -   Accept two iterables
        -   x-coordinates
        -   heights

```python
data = np.random.randint(0, 20, 10)
plt.bar(np.arange(len(data)), data);
```

```python
plt.bar(np.arange(len(data)), data)
plt.plot(data, c='y')
```

```python
precip_2018 = [3.92,
             1.7000000000000002,
             4.800000000000001,
             5.35,
             5.119999999999999,
             1.98,
             4.919999999999999,
             6.619999999999999,
             7.9700000000000015,
             4.579999999999999,
             7.1,
             6.2299999999999995]
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
       'Oct', 'Nov', 'Dec'])
plt.bar(np.arange(len(precip_2018)), precip_2018)
plt.xticks(ticks=np.arange(len(precip_2018)), labels=months);
```

### <font color=lightGreen>Horizontal bar charts</font>
-   Generated by the `barh()` function
-   Length of the bar represents the magnitude of the values

```python
plt.barh(np.arange(len(precip_2018)), precip_2018)
plt.yticks(ticks=np.arange(len(precip_2018)), labels=months);
```

### <font color=lightGreen>Multiple datasets</font>
<font color=mediumTurquoise>Stacked</font> bar chars

```python

func = np.random.rand
dataset_1 = func(12)
dataset_2 = func(12)
dataset_3 = func(12)
```

```python
plt.bar(np.arange(len(dataset_3)), dataset_3)
plt.bar(np.arange(len(dataset_2)), dataset_2)
plt.bar(np.arange(len(dataset_1)), dataset_1);
```

This way bars are one on top to another, in order o get stacked bars some adjustments have to be done to provide the needed offset, as follows:

The bar function has a keyword argument, `bottom`, which is basically the offset of the Y-axis for the dataset.
Since I want dataset_3 at the bottom of the stack, its offset is 0, which is the default, so I can leave that function call alone.
The dataset_2 data I want to offset by the heights of the dataset_3 data.
So I'll give that to the bottom keyword argument.
The dataset_1 data needs to be offset by both the dataset_2 and dataset_3 heights

```python
plt.bar(np.arange(len(dataset_3)), dataset_3)
plt.bar(np.arange(len(dataset_2)), dataset_2, bottom=dataset_3)
plt.bar(np.arange(len(dataset_1)), dataset_1, bottom=dataset_3 + dataset_2)
plt.xticks(ticks=np.arange(len(precip_2018)), labels=months);
```

To stack a horizontal bar chart.
There are just three changes to make.
First, I must change the bar function calls to `barh`.
The `bottom` keyword argument no longer makes sense, so it has been changed to `left`.
And finally, the call to `xticks` must become `yticks`

```python
plt.barh(np.arange(len(dataset_3)), dataset_3)
plt.barh(np.arange(len(dataset_2)), dataset_2, left=dataset_3)
plt.barh(np.arange(len(dataset_1)), dataset_1, left=dataset_3 + dataset_2)
plt.yticks(ticks=np.arange(len(precip_2018)), labels=months);
```

<font color=mediumTurquoise>Grouped</font> bar charts

```python
group_width = 0.9
plt.bar(np.arange(len(dataset_3)) - group_width / 3, dataset_3, group_width / 3)
plt.bar(np.arange(len(dataset_2)), dataset_2, group_width / 3)
plt.bar(np.arange(len(dataset_1)) + group_width / 3, dataset_1, group_width / 3)
plt.xticks(ticks=np.arange(len(precip_2018)), labels=months);
```
