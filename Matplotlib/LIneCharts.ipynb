### <font color=lightGreen>Demo - line Chart</font>

```python
precip_2018 = [3.92,
             1.7000000000000002,
             4.800000000000001,
             5.35,
             5.119999999999999,
             1.98,
             4.919999999999999,
             6.619999999999999,
             7.9700000000000015,
             4.579999999999999,
             7.1,
             6.2299999999999995]
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
       'Oct', 'Nov', 'Dec'])
```

```python
plt.plot(precip_2018)
plt.xticks(ticks=np.arange(len(precip_2018)), labels=months);
```

Passing the ticks keyword argument an array with just a few locations, only the positions in the ticks keyword are displayed

```python
plt.plot(precip_2018)
plt.xticks(ticks=np.array([1, 2, 8, 10]), labels=months);
```

It would also make sense to show the months associated with the positions instead of just the first four

```python
plt.plot(precip_2018)
plt.xticks(ticks=np.array([1, 2, 8, 10]), labels=months[[1, 2, 8, 10]]);
```

The `np.argmin` and `np.argmax` functions will find the minimum and maximum precipitation for the year.
Now I'll plot the totals again. But I'll only label the minimum and the maximum.

```python
low_precip = np.argmin(precip_2018)
high_precip = np.argmax(precip_2018)
plt.plot(precip_2018)
plt.xticks(ticks=np.array([low_precip, high_precip]), labels=months[[low_precip, high_precip]]);
```

To label just the maximum value, I simply have to change labels to an array with only one value

```python
plt.plot(precip_2018)
plt.xticks(ticks=np.array([high_precip]), labels=months[[high_precip]]);
```

This is a shortcut if you want to display no ticks at all

```python
plt.plot(precip_2018)
plt.xticks(ticks=[]);
```

This also works for the Y-axis as well with a function named creatively `yticks`.
So if I wanted to show the ticks along the Y-axis spaced in quarter inches instead of whole inches,
I'll need to write a little code to determine the range between the months with the highest and lowest precipitation and then create an array with those bounds evenly spaced at 0.25 to use as the ticks for the Y-axis

```python
import math
lower = math.floor(precip_2018[low_precip])
upper = math.ceil(precip_2018[high_precip])
precip_ticks = np.arange(lower, upper, 0.25)
plt.plot(precip_2018)
plt.xticks(ticks=np.arange(len(precip_2018)), labels=months)
plt.yticks(ticks=precip_ticks);
```
