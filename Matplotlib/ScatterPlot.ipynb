## <center><font color=tomato>Scatter Plot</font></center>
-   Display individual points in two dimensions
-   Order is not significant
-   Scatter plots shows relationships between the points
-   Generated by the `scatter()` function
-   Accepts two iterables
    -   First is the x-axis
    -   Second is the y-axis

comparison with plot...

```python
x = np.linspace(start=-10, stop=10, num=50)
y = x ** 3
plt.plot(x,y);
```

```python
plt.scatter(x,y);
```

Regardless of the order that the points are given to the scatter plot, the result will be the same.
And here is the proof of what I mean. We just saw this scatter plot of the equation X to the third power.
If I zip the X and Y coordinates, thus using tuples to join them,
I can then randomly rearrange the order of those tuples with the shuffle function found in the random module in the Python-centered library

```python
import random
points = list(zip(x, y))
random.shuffle(points)
points = np.array(points)
plt.scatter(points[:,0], points[:,1])
```

The `make_blobs` function from the `Scikit-learn` module will generate points that form clusters or groups.
The return value is a tuple, and the first element is a list of points.

```python
from sklearn.datasets import make_blobs
data = make_blobs()
points = data[0]
x, y = points[:, 0], points[:, 1]
plt.scatter(x, y);
```

The other element in the data tuple is an array of categories.
There's a category for each set of points. I can assign a color to each category using the `c` keyword and passing it the array of categories.
This actually adds a third dimension to the scatter plot, but it's not a spatial dimension

```python
data = make_blobs()
points = data[0]
x, y = points[:, 0], points[:, 1]
plt.scatter(x, y, c=data[1]);
```

Actually `make_blobs` function has the option to provide the number of categories

```python
data = make_blobs(n_samples=200, centers=5)
points = data[0]
x, y = points[:, 0], points[:, 1]
plt.scatter(x, y, c=data[1]);
```
